### Question 5 : What is the maximum & minimum 'average_price' per year in england ?

```python
# Analyzing data

df.head(2)
```

```python
# Distinct areas

df['area'].unique()
```

```python
# Records of england

df[df['area'] == 'england']
```

```python
# New column year

df['year'] = df['date'].dt.year
```

```python
# Records of england Maximum average price

df[df['area'] == 'england'].groupby(['year'])['average_price'].max()
```

```python
# Records of england Minimum average price

df[df['area'] == 'england'].groupby(['year'])['average_price'].min()
```
