# Exploratory Data Analysis

## Gender

```python
# Checking columns name

df.columns
```

```python
# Visual representation of gender

gen_represent = sns.countplot(x = 'Gender', data = df)
sns.set(rc = {'figure.figsize': (3,7)})
```

```python
gen_represent = sns.countplot(x = 'Gender', data = df, hue = 'Gender')

for bars in gen_represent.containers :
    gen_represent.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
amt_represent = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y = 'Amount', data = amt_represent, hue = 'Gender')
```

From the above graphs we can see that most of the buyers are females and even the purchasing power of female is greater than man.

## State

```python
df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
```

```python
state_order = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':( 20, 5)})
sns.barplot(x = 'State', y = 'Orders', data = state_order, hue = 'State')
```

```python
state_order = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':( 20, 5)})
state_order_represent = sns.barplot(x = 'State', y = 'Orders', data = state_order, hue = 'State')

for bars in state_order_represent.containers :
    state_order_represent.bar_label(bars)
```

```python
state_amt = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
state_amt
```

```python
state_amt = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
state_amt_represent = sns.barplot(x = 'State', y = 'Amount', data = state_amt, hue = 'State')

for bars in state_amt_represent.containers :
    state_amt_represent.bar_label(bars)
```

From the above graphs we can see that most of orders and most amount paid are from the states of Uttar pradesh, Maharashtra, karnataka

## Marital Status

```python
martial_count = sns.countplot(x = 'Marital_Status', data = df, hue = 'Marital_Status')
sns.set(rc = {'figure.figsize': (7,3)})

for bars in martial_count.containers:
    martial_count.bar_label(bars)
```

```python
gender_martial_count = df.groupby(['Marital_Status'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
gender_martial_count
```

```python
gender_martial_count = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Marital_Status', y = 'Amount', data = gender_martial_count, hue = 'Gender')
```

From the above charts we can see that most of the buyers are unmarried womens and they have high purchasing power.

## Occupation

## Product Category

```python
# Checking columns names

df.columns
```

```python
# Visualizing Product category

sns.set(rc = { 'figure.figsize':(20,5)} )
product_catg = sns.countplot( x = 'Product_Category', data = df, hue = 'Product_Category')


for bars in product_catg.containers:
    product_catg.bar_label(bars)
```

```python
# Analyzing the data 

product_catg_amt = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
product_catg_amt
```

```python
# Visualizing the sales by category

sns.set(rc = { 'figure.figsize':(20,5)} )

product_catg_amt_represent = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(x = 'Product_Category', y = 'Amount', data = product_catg_amt_represent, hue = 'Product_Category')
```

From the above graphs we can see that most of the sold product are from Food, clothings, Electronic.
