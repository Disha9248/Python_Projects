# Covid 19 Analysis

### Question 1 : Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
# Analyzing data

df.head(3)
```

```python
# Number of confirmed, Deaths and Recovered cases in each region

df.groupby(['Region'])[['Confirmed','Deaths','Recovered']].sum()
```

### Question 2 : Remove all the records where the Confirmed Cases is Less Than 10.

### Question 3 : In which Region, maximum number of Confirmed cases were recorded ?

```python
# Analyzing data

df.head(2)
```

```python
# Region and maximum confirmed cases

df.groupby(['Region'])['Confirmed'].sum().sort_values( ascending = False).head(10)
```

### Question 4 : In which Region, minimum number of Deaths cases were recorded ?

```python
# Analyzing data

df.head(2)
```

```python
# Region and minimum deaths cases

df.groupby(['Region'])['Deaths'].sum().sort_values( ascending = True).head(15)
```
